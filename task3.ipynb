## Linear filtering by convolution (mean filter)

**<a id="#">Task 3.1.1.</a>** Load the image `data/lena.png` and show it:

```python
img = plt.imread('data/lena.png')
plt.imshow(img)
```

**<a id="#">Task 3.1.2.</a>** Implement a mean filter:

```python
def meanfilter(img, size):
    result = numpy.zeros(img.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(img.shape[0]-2*n, img.shape[1]-2*n):

        neighborhood_sum = 0
        for q in numpy.ndindex(size, size):
            neighborhood_sum += img[p[0]+q[0],p[1]+q[1]] 
        result[p[0]+n,p[1]+n] = (1/(size*size)) * neighborhood_sum 
    return result
```

## Non-linear filtering

**<a id="#">Task 3.2.1.</a>** Implement a median filter:

```python
def medianfilter(img, size):
    result = numpy.zeros(img.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(img.shape[0]-2*n, img.shape[1]-2*n):
        neighborhood_sum = 0
        for q in numpy.ndindex(size, size):
            neighborhood_sum += img[p[0]+q[0],p[1]+q[1]] 
        result[p[0]+n,p[1]+n] = (1/size*size) * neighborhood_sum 
    return result
```
